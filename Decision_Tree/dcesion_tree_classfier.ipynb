```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
print(iris["target"])
```

```python
X=pd.DataFrame(iris["data"],columns=["sepal_length","sepal_wedth","petal_lenght","petal_width"])
X.head()
```

```python
Y=pd.DataFrame(iris["target"])
Y.head()
```

split the data for tarin and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

build the model

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

visulaize the decsiuon tree


```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

apply the post prunning

```python
dt_p=DecisionTreeClassifier(max_depth=2)
dt_p.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_p,filled=True)
```

```python
y_pred_dt=dt.predict(X_test)
y_pred_dt_p=dt_p.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc_dt=accuracy_score(y_pred_dt,y_test)
acc_dt_p=accuracy_score(y_pred_dt_p,y_test)
```

```python
print("accuracy_score without post pruning",acc_dt)
print("accuracy_score with post prunning",acc_dt_p)
```
